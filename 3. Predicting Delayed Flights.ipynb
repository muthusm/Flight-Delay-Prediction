```
df = sqlContext.table("FLIGHTDELAY")
```

```
df.count()
```

```
## Filtering df with only non cancelled and non diverted flights
df = df.filter((df.Cancelled == 0) & (df.Diverted == 0))
```

```
df.count()
```

```
# DELAY - Flights with Arrival delay greater or equal to 15 min
df = df.withColumn('Delayed', (df.ArrDelay >=15).cast('int')) 

## TIME SLOT by departure time (using SQL Context)
sqlContext.registerDataFrameAsTable(df, "FLIGHTDELAY_1")

df = sqlContext.sql("select *, case \
               when CRSDepTime <= 800 then 1 \
               when 800 < CRSDepTime and CRSDepTime <= 1200 then 2 \
               when 1200 < CRSDepTime and CRSDepTime <= 1600 then 3 \
               when 1600 < CRSDepTime and CRSDepTime <= 2100 then 4 \
               else 1 end as TimeSlot \
               from FLIGHTDELAY_1")
```

```
df.count()
```

```
display(df)
```

```
# Indexing "UniqueCarrier" (from string to numeric)

from pyspark.ml.feature import StringIndexer

indexer = StringIndexer(inputCol='Reporting_Airline',outputCol='IndexUniqueCarrier') 
indexed_data=indexer.fit(df).transform(df)

indexed_data.groupBy('Reporting_Airline','IndexUniqueCarrier').count().sort('IndexUniqueCarrier').show()
```

```
indexed_data = indexed_data.na.drop(subset=['DepDelay', 'Delayed']) \
                              .select('DepDelay',
                                   'Distance', 
                                   'DayOfWeek',
                                   'IndexUniqueCarrier', 
                                   'TimeSlot',
                                   'Delayed'
                                  )

indexed_data.limit(10).toPandas()
```

```
indexed_data.count()
```

```
from pyspark.ml.feature import VectorAssembler, StringIndexer  
from pyspark.sql.functions import col

assembler = VectorAssembler(  
                        inputCols=['DepDelay',  'Distance',  'DayOfWeek',  'IndexUniqueCarrier',  'TimeSlot'],  
                        outputCol='features'
)

final_data = assembler.transform(indexed_data).select(col("Delayed").cast('double').alias("label"),'features')  
final_data.show(5)
```

```
stringIndexer = StringIndexer(inputCol = 'label', outputCol = 'isDelayed')  
sI = stringIndexer.fit(final_data)  
final_data = sI.transform(final_data)  
final_data = final_data.select('isDelayed','features') 
```

```
final_data.show(5)
```

```
train_data, test_data = final_data.randomSplit([0.7, 0.3], seed=123)
```

```
train_data.count()
```

Creating a Random Forest and predict delayed flights

```
from pyspark.ml.classification import RandomForestClassifier
from time import time

start_time = time()

# Train a RandomForest model
rf = RandomForestClassifier(labelCol="isDelayed", 
     featuresCol="features", 
     numTrees=500,
     maxDepth=3, 
     seed = 1,
     featureSubsetStrategy="sqrt",
     impurity='gini') 

model = rf.fit(train_data)

end_time = time()
elapsed_time = end_time - start_time
print("Time to train model: %.3f seconds" % elapsed_time)
```

Prediction and metrics

```
pred = model.transform(test_data)

cm = pred.groupBy('isDelayed','prediction').count().sort('isDelayed','prediction')
print("Confusion matrix")
cm.show()

TN = cm.filter((cm.isDelayed == 0) & (cm.prediction == 0)).collect()[0][2] #True negative
FP = cm.filter((cm.isDelayed == 0) & (cm.prediction == 1)).collect()[0][2] #False positive
FN = cm.filter((cm.isDelayed == 1) & (cm.prediction == 0)).collect()[0][2] #Flase negative
TP = cm.filter((cm.isDelayed == 1) & (cm.prediction == 1)).collect()[0][2] #True positive

N=TN+FP+TP+FN

Prev = (TP + FN) / N  
Sens = TPR = Recall = TP / (TP + FN) 
Esp  = TN / (TN + FP) #= (1 - FPR) 
Precision = PPV = TP / (TP + FP) 
Acc = (TP+TN) / N  

print("Metrics:")
print('Prevalence=',round(Prev,2))
print('Sensitivity=',round(Sens,2))
print('Especificity=',round(Esp,2))
print('Recall=',round(Recall,2))
print('Precision=',round(Precision,2))
print('Accuracy=',round(Acc,2))
```

ROC curve

```
from pyspark.sql.functions import udf
from pyspark.sql.types import FloatType

secondelement=udf(lambda v:float(v[1]),FloatType())
pred2=pred.withColumn('probs',secondelement('probability'))

import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve
import seaborn as sns

pd5 = pred2.sample(False,0.1).select('isDelayed','probs').toPandas()

fpr, tpr, _ = roc_curve( pd5['isDelayed'], pd5['probs'])

plt.plot(fpr, tpr)
plt.plot([0, 1], [0, 1], color='navy', linestyle='--')
plt.xlabel('1 - Esp (FPR)')
plt.ylabel('Sens (TPR)')
plt.title('ROC Curve')
plt.show()
```
